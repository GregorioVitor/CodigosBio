```python
from Bio import SeqIO
import pandas as pd
from Bio.SeqRecord import SeqRecord
import re
```

```python
auxDrerio=[]
for record in SeqIO.parse('RefSeq/Drerio/GCF_000002035.6_GRCz11_rna_from_genomic.fna',"fasta"):
    print(record.description)
    print()
    teste = record.description.split('gbkey=')
    auxDrerio.append(re.split("]", teste[1])[0])
    
seqDrerio=[]
for i in range(len(auxDrerio)):
    seqDrerio.append('Seq_'+str(i+1))
```

```python
dfDrerio = pd.DataFrame(columns=['Seq','ncRNA'])
dfDrerio.ncRNA=auxDrerio
dfDrerio.Seq = seqDrerio
dfDrerio
```

```python
dfDrerioncRNA = dfDrerio[dfDrerio['ncRNA']!='mRNA']
dfDrerioncRNA = dfDrerioncRNA.replace('precursor_RNA', 'miRNA') 
dfDrerioncRNA
```

```python
from Bio import SeqIO
import pandas as pd
from Bio.SeqRecord import SeqRecord
import re
```

```python
auxDmelanogaster=[]
for record in SeqIO.parse('RefSeq_mRNA/FASTAs/New_Dmelanogaster_3.fasta',"fasta"):
    auxDmelanogaster.append(record.id)
    
seqDmelanogaster=[]
for i in range(len(auxDmelanogaster)):
    seqDmelanogaster.append(i+1)
dfDmelanogaster = pd.DataFrame(columns=['Index','seqRNA'])
dfDmelanogaster.seqRNA=auxDmelanogaster
dfDmelanogaster.Index = seqDmelanogaster

findDmelanogaster = pd.read_csv('RefSeq_mRNA/FindNonCoding/Dmelanogaster/New_Dmelanogaster_3.csv', sep=',')
findDmelanogaster['Index'] = findDmelanogaster.Index.astype('object')

resultadoDmelanogaster = pd.merge(dfDmelanogaster, findDmelanogaster, how='inner', on = 'Index')
resultadoDmelanogaster.to_csv('RefSeq_mRNA/FindNonCoding/Dmelanogaster/Dmelanogaster_3_arrumado.csv', index=False)

auxDmel=[]
for record in SeqIO.parse('RefSeq_mRNA/FASTAs/Dmelanogaster.fna',"fasta"):
    string = record.description.replace('gbkey=', 'ncRNA_class=')
    teste = re.split("ncRNA_class=", string)
    auxDmel.append(re.split("]", teste[1])[0])
    
seqDmel=[]
for i in range(len(auxDmel)):
    seqDmel.append("Seq_"+str(i+1))

dfDmel = pd.DataFrame(columns=['Seq','ncRNA'])
dfDmel.ncRNA=auxDmel
dfDmel.Seq = seqDmel


findDmelRef = pd.read_csv('RefSeq_mRNA/FindNonCoding/Dmelanogaster/Dmelanogaster_3_arrumado.csv')
findDmelRef['ncOriginal']=0
findDmelRef['result']=0
for i in range(len(findDmelRef)):
    res = dfDmel[dfDmel['Seq']==findDmelRef['seqRNA'][i]]
   
    if res['ncRNA'].str.cat().lower() == findDmelRef['Gene'][i].lower():
        findDmelRef['ncOriginal'][i] = res['ncRNA'].str.cat().lower()
        findDmelRef['result'][i]=1
    else:
        findDmelRef['ncOriginal'][i] = res['ncRNA'].str.cat().lower()
        findDmelRef['result'][i]=0
findDmelRef.to_csv('RefSeq_mRNA/FindNonCoding/Dmelanogaster/Dmelanogaster_3_arrumado.csv')
```

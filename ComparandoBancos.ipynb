```python
from Bio import SeqIO
import pandas as pd
import pickle
import os
```

```python
if os.path.exists("RNAcentral_active.pkl"):
    infile = open("RNAcentral_active.pkl",'rb')
    dici = pickle.load(infile)
    infile.close()
else: 
    dici={}
    for seq_record in SeqIO.parse("rnacentral_active.fasta", "fasta"):
        frase = seq_record.description.split()
        dici[frase[0]]=frase[1]
    with open("RNAcentral_active.pkl", "wb") as tf:
        pickle.dump(dici,tf)
```

```python
from Bio import SeqIO
import pandas as pd
from Bio.SeqRecord import SeqRecord
import re
```

```python
auxDmel=[]
for record in SeqIO.parse('RefSeq/Drerio/GCF_000002035.6_GRCz11_rna_from_genomic.fna',"fasta"):
    print(record.description)
    print()
    teste = record.description.split('gbkey=')
    auxDmel.append(re.split("]", teste[1])[0])
```

```python
seqDmel=[]
for i in range(len(auxDmel)):
    seqDmel.append('Seq#'+str(i+1))
```

```python
dfDmel = pd.DataFrame(columns=['Seq','ncRNA'])
dfDmel.ncRNA=auxDmel
dfDmel.Seq = seqDmel
dfDmel
```

```python
dfDmel_ncRNA= dfDmel[dfDmel['ncRNA']!='mRNA']
dfDmel_ncRNA
```

```python
dfDmel_mRNA = dfDmel[dfDmel['ncRNA']=='mRNA']
dfDmel_mRNA
```

```python
len(dfDmel_ncRNA)
```

```python
mRNA_aleatorios = dfDmel_mRNA.sample(n=18326)
mRNA_aleatorios
```

```python
mRNA_aleatorios.drop_duplicates()
```

```python
tabelaFinal = pd.concat([mRNA_aleatorios, dfDmel_ncRNA])
tabelaFinal = tabelaFinal.sort_values(['Seq'])
tabelaFinal = tabelaFinal.reset_index(drop=True)
tabelaFinal
```

```python
frags=[]
n=1
for record in SeqIO.parse('RefSeq/Drerio/GCF_000002035.6_GRCz11_rna_from_genomic.fna', 'fasta'):
    frag = record.seq
    gravar = SeqRecord(frag, 'Seq#'+str(n), "", "")
    frags.append(gravar)
    n=n+1
SeqIO.write(frags, 'RefSeq/Drerio/New_GCF_000002035.6_GRCz11_rna_from_genomic.fna', 'fasta')
```

```python
frags=[]
for i, j in tabelaFinal.iterrows():
    for record in SeqIO.parse('RefSeq/Drerio/New_GCF_000002035.6_GRCz11_rna_from_genomic.fna', 'fasta'):
        if tabelaFinal.loc[i,'Seq'] == record.id:
                    frag = record.seq
                    gravar = SeqRecord(frag, record.id, "", "")
                    frags.append(gravar)
SeqIO.write(frags, 'RefSeq/Drerio/FinalAleatorio_GCF_000002035.6_GRCz11_rna_from_genomic.fna', 'fasta')
```

```python
from Bio import SeqIO
import pandas as pd
import pickle
import os
```

```python
if os.path.exists("RNAcentral_active.pkl"):
    infile = open("RNAcentral_active.pkl",'rb')
    dici = pickle.load(infile)
    infile.close()
else: 
    dici={}
    for seq_record in SeqIO.parse("rnacentral_active.fasta", "fasta"):
        frase = seq_record.description.split()
        dici[frase[0]]=frase[1]
    with open("RNAcentral_active.pkl", "wb") as tf:
        pickle.dump(dici,tf)
```

```python
blast = "blastn_TriTrypDB-54_LdonovaniBPK282A1.tsv"
```

```python
arq = pd.read_csv(blast, sep="\t", names=["qseqid", "sseqid","pident","length", "mismatch", "gapopen", "qstart" ,"qend", "sstart", "send",'fita', "evalue" ,"bitscore" ,"qlen", "slen" ])
```

```python
arq = arq.query('pident>= 95' )
arq
```

```python
arq['type'] = 'NaN'
```

```python
arq['type']=arq['qseqid'].map(dici)
```

```python
arq.to_csv("final_blastn_TriTrypDB-54_LdonovaniBPK282A1.csv", index=False)
```

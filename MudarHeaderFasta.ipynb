```python
from Bio import SeqIO
from Bio.SeqRecord import SeqRecord
from Bio.Seq import Seq
```

```python
# Muda todos os cabeçalho para chr1
frags=[]
for record in SeqIO.parse('teste/transcript.fa', "fasta"):
    gravar = SeqRecord(record.seq, "chr1","","")
    frags.append(gravar)
    SeqIO.write(frags, "teste/teste.fa", "fasta")
```

```python
for record in SeqIO.parse('teste/teste.fa', "fasta"):
    print(record)
```

```python
# Muda o cabeçalho para chr1, chr2, chr3 ... e salva o cabeçalho original junto como descrição
i=1
frags=[]
for record in SeqIO.parse('teste/transcript.fa', "fasta"):
    gravar = SeqRecord(record.seq, "chr" +str(i),"",record.id)
    frags.append(gravar)
    SeqIO.write(frags, "teste/teste2.fa", "fasta")
    i+=1
```

```python
for record in SeqIO.parse('teste/teste2.fa', "fasta"):
    print(record)
```

```python
#Salva todas as sequencias em uma sequencia só
frags=[]
for record in SeqIO.parse('teste/transcript.fa', "fasta"):
    frags.append(str(record.seq))

teste = "".join(frags)
gravar = SeqRecord(Seq(teste), "chr1" ,"","")
SeqIO.write(gravar, "teste/teste3.fa", "fasta")
```

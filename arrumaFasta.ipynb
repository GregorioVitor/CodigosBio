```python
from Bio import SeqIO
import pandas as pd
from Bio.SeqRecord import SeqRecord
import re
import argparse
import os


aux=[]
for record in SeqIO.parse('RefSeq_mRNA/FASTAs/Dmelanogaster.fna',"fasta"):
    teste = record.description.split('gbkey=')
    aux.append(re.split("]", teste[1])[0])

seq=[]
for i in range(len(aux)):
    seq.append('Seq_'+str(i+1))

df = pd.DataFrame(columns=['Seq','ncRNA'])
df.ncRNA=aux
df.Seq = seq

df_ncRNA= df[df['ncRNA']!='mRNA']
df_mRNA= df[df['ncRNA']=='mRNA']

num = len(df_ncRNA)
df_mRNA
```

```python
df_ncRNA
```

```python
mRNA_aleatorio1 = df_mRNA.sample(n=num)
aux2= df_mRNA[~df_mRNA.isin(mRNA_aleatorio1)].dropna()

mRNA_aleatorio2 = aux2.sample(n=num)
aux3= df_mRNA[~aux2.isin(mRNA_aleatorio2)].dropna()

mRNA_aleatorio3 = aux3.sample(n=num)
aux4= df_mRNA[~aux3.isin(mRNA_aleatorio3)].dropna()

mRNA_aleatorio4 = aux4.sample(n=num)
aux5= df_mRNA[~aux4.isin(mRNA_aleatorio4)].dropna()

mRNA_aleatorio5 = aux5.sample(n=num)
mRNA_aleatorio5
```

```python
tabelaFinal1 = pd.concat([mRNA_aleatorio1, df_ncRNA]).sort_values(['Seq']).reset_index(drop=True)
tabelaFinal2 = pd.concat([mRNA_aleatorio2, df_ncRNA]).sort_values(['Seq']).reset_index(drop=True)
tabelaFinal3 = pd.concat([mRNA_aleatorio3, df_ncRNA]).sort_values(['Seq']).reset_index(drop=True)
tabelaFinal4 = pd.concat([mRNA_aleatorio4, df_ncRNA]).sort_values(['Seq']).reset_index(drop=True)
tabelaFinal5 = pd.concat([mRNA_aleatorio5, df_ncRNA]).sort_values(['Seq']).reset_index(drop=True)
```

```python
def escreve_fasta(tabelaFinal, aux):
    ids = list(tabelaFinal['Seq'])
    i=1
    frags=[]
    for record in SeqIO.parse('RefSeq_mRNA/FASTAs/New_Dmelanogaster.fasta', "fasta"):
            for i in range(len(ids)):
                if record.id == ids[i]:
                    frag = record.seq
                    gravar = SeqRecord(frag, ids[i], '', '')
                    frags.append(gravar)
    SeqIO.write(frags, 'RefSeq_mRNA/FASTAs/New_Dmelanogaster_'+str(aux)+'.fasta', "fasta")
```

```python
escreve_fasta(tabelaFinal5, 5)
escreve_fasta(tabelaFinal4, 4)
escreve_fasta(tabelaFinal3, 3)
escreve_fasta(tabelaFinal2, 2)
escreve_fasta(tabelaFinal1, 1)
```

```python
auxDmel=[]
for record in SeqIO.parse('RefSeq_mRNA/FASTAs/Scerevisiae.fna',"fasta"):
    string = record.description.replace('gbkey=', 'ncRNA_class=')
    teste = re.split("ncRNA_class=", string)
    
    if re.split("]", teste[1])[0] == 'precursor_RNA':
        teste3 = re.split("gene=", string)
        if re.split("-", teste3[1])[0] != 'mir':
            auxDmel.append('precursor_RNA')
        else: 
            auxDmel.append('miRNA')
    elif re.split("]", teste[1])[0] == 'other':
        teste2 = re.split("gene=", string)
        auxDmel.append(re.split(":", teste2[1])[0])
    elif re.split("]", teste[1])[0] == 'ncRNA':
        teste4 = re.split("gene=", string)
        auxDmel.append(re.split(":", teste4[1])[0])
    else:
        auxDmel.append(re.split("]", teste[1])[0])
```

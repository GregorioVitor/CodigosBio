```python
from Bio import SeqIO
import pandas as pd
from Bio.SeqRecord import SeqRecord
import re
```

```python
fasta = 'Dmelanogaster/dmel-all-ncRNA-r6.47.fasta'
```

```python
df = pd.DataFrame(columns=['Seq', 'ncRNA'])
```

```python
aux=[]
for record in SeqIO.parse(fasta,"fasta"):
    print(record.description)
    print()
    teste = record.description.split('name=')
    aux.append(re.split("[:;]", teste[1])[0])
```

```python
df.ncRNA=aux
```

```python
seq=[]
for i in range(3053):
    seq.append('Seq_'+str(i))
```

```python
df.Seq = seq
```

```python
df
```

```python
gff = 'Dmelanogaster/Dmelanogaster_ResultadoFinal.gff'
```

```python
gffVitor = pd.read_csv(gff, skiprows=1, names = ['Seq', 'tool', 'ncRNA', 'start', 'end', 'score', 'strand', '.', 'id'], sep='\t')
gffVitor = gffVitor[['Seq', 'ncRNA']]
gffVitor
```

```python
gffVitor['ncOriginal']=0
gffVitor['result']=0

for i in range(len(gffVitor)):
    res = df[df['Seq']==gffVitor['Seq'][i]]
    print(res['ncRNA'].str.cat().lower())
    print(gffVitor['ncRNA'][i].lower())
    print()
    
    if res['ncRNA'].str.cat().lower() == gffVitor['ncRNA'][i].lower():
        gffVitor['ncOriginal'][i] = res['ncRNA'].str.cat().lower()
        gffVitor['result'][i]=1
    else:
        gffVitor['ncOriginal'][i] = res['ncRNA'].str.cat().lower()
        gffVitor['result'][i]=0
```
